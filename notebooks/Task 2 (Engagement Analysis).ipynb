Read in Data

```python
df = pd.read_csv('telco.csv')
```

```python
df.head()
```

Drop rows with missing values in MSISDN/Number and Bearer Id

```python
df = df[df['MSISDN/Number'].notna()]
```

```python
df = df[df['Bearer Id'].notna()]
```

```python
df = df.reset_index()
```

Sort columns to Numerical and Categorical

```python
num_cols = []
categorical_cols = []
```

```python
def sort_cols(data):


    for i in data.columns:
        if data[i].dtypes == "object":
            categorical_cols.append(i)
        else :
            num_cols.append(i)
```

```python
sort_cols(df)
Df = df.copy()
```

Scale Data using Standard Scaler

```python
df[num_cols] = StandardScaler().fit_transform(df[num_cols])
```

Fix missing values

```python
def fix_cat_cols(data,cat_columns):
    for i in cat_columns:
        data[i] = data[i].fillna(data[i].mode()[0])
```

```python
fix_cat_cols(df,categorical_cols)
```

```python
def fix_num_cols(data,num_columns):
    for i in num_columns:
        data[i] = data[i].fillna(data[i].mean())
```

```python
fix_num_cols(df,num_cols)
```

Sort using Kmeans clustering to K different clusters

```python
kmeans_kwargs ={
    'init': "random",
    'n_init': 10,
    'max_iter': 300,
    'random_state': 42
}
```

```python
sse = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, **kmeans_kwargs)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
```

Check for optimized number of clusters using the elbow method then interprete the findings.

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1,11),sse)
plt.xticks(range(1,11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

- The Optimized value for k is 2

- Deduced a steep slope k is from 1 till 2, then the slope starts to flatten out from k = 3 till 10 where the slope stopped
